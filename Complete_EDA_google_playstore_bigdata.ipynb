### **7. Let fix `size` column**

- Let's clean the `size` column first

```python
# Display unique values in the 'size' column
df['size'].unique()
```

-  There are several uniques values in the `size` column, we have to first make the unit into one common unit from M and G to K, and then remove the `G` ,`M`, and `K`, and from the values and convert them into numeric data type. `Varies with device` into NaN
-  Here, we also notice the some values have `,` which make a hardle to convert them into bytes so first we remove these commas also

```python
# find the values in size column which has ',' in it
# we have some NaN values 
# So, Use isna() to exclude NaN
df['size'].loc[(df['size'].str.contains(',')) & (~df['size'].isna())].value_counts().sum()
```

```python
# find the values in size column which has 'G' in it
df['size'].loc[(df['size'].str.contains('G')) & (~df['size'].isna())].value_counts().sum()
```

```python
# find the values in size column which has 'M' in it
df['size'].loc[(df['size'].str.contains('M')) & (~df['size'].isna())].value_counts().sum()
```

```python
# find the values in size column which has 'k' in it
df['size'].loc[(df['size'].str.contains('k')) & (~df['size'].isna())].value_counts().sum()
```

```python
# find the values in size column which has 'Varies with device' in it
df['size'].loc[(df['size'].str.contains('Varies with device')) & (~df['size'].isna())].value_counts().sum()
```

> Let's convert the `G` and `M` units into `k` and then remove the `G`, `M` and `k` from the values and convert them into numeric data type.

```python
# Define a function to convert the 'size' column to numeric
def convert_size(size_value):
    # add function details here
    '''
    This function will convert the size column to numeric by multiplying the values with 1024 if it has 'M' in it and 1024*1024 if it has 'G' in it
    If 'Varies with device' is present, it returns NaN.
    '''

    if isinstance(size_value, str):
        # Remove commas from the size values
        size_value= size_value.replace(',','')
        if 'k' in size_value:
            # Convert to float and remove 'k'
            return float(size_value.replace('k', ''))
        elif 'M' in size_value:
            # Convert to float, remove 'M', and multiply by 1024
            return float(size_value.replace('M', '')) * 1024
        elif 'G' in size_value:
            # Convert to float, remove 'G', and multiply by 1024*1024
            return float(size_value.replace('G', '')) * 1024 * 1024
        elif 'Varies with device' in size_value:
            # If 'Varies with device' is present, return NaN
            return np.nan
    return size_value

# Apply the conversion function to the 'size' column in the DataFrame
df['size'] = df['size'].apply(convert_size)
```

```python
# rename the column name 'Size' to 'Size_in_bytes'
df.rename(columns={'size': 'size_in_KB'}, inplace=True)
```

```python
# making a new column called 'Size in Mb' which will have the size in MB
df['size_in_MB'] = df['size_in_KB'].apply(lambda x: x/1024)
```

### **8. Let's fix `installs` column**

- Let's clean the `installs` column

```python
# Count the occurrences of each unique value in the 'installs' column using value_counts()
df['installs'].value_counts()
```

```python
# Count the total number of occurrences of each unique value in the 'installs' column using value_counts()
df['installs'].value_counts().sum()
```

```python
# Count the occurrences of the plus sign character ('+') in the 'installs' column using str.contains()
df['installs'].str.contains('\\+').sum()
```

```python
# Count the occurrences of the comma character (',') in the 'installs' column using str.contains()
df['installs'].str.contains('\\,').sum()
```

```python
# remove the plus sign from install column and convert it to numeric
df['installs'] = df['installs'].apply(lambda x: x.replace('+', '') if '+' in str(x) else x)

# also remove the commas from the install column
df['installs'] = df['installs'].apply(lambda x: x.replace(',', '') if ',' in str(x) else x)

# Or we can use this method also
# df['Installs_tmp'] = df['Installs_tmp'].str.replace(',', '').str.replace('+', '')
```

> In this, the errors='coerce' parameter will replace non-convertible values with NaN, and the .astype('Int64') ensures that the column remains of nullable integer type (pandas nullable integer type that supports NaN). This way, missing values will be represented as NaN after conversion. Adjust the column name ('your_column') accordingly based on your actual DataFrame.

```python
# Convert the 'installs' column to numeric, coercing errors to NaN using pd.to_numeric()
# Then, convert the numeric values to Nullable integer type (Int64) using astype('Int64')
df['installs'] = pd.to_numeric(df['installs'], errors='coerce').astype('Int64')
```

### **9. Let deal with `duplicated` and `Missing values`**

- Lets check if we have duplicated or not

```python
# Count the number of duplicated rows in the DataFrame using df.duplicated()
df.duplicated().sum()
```

```python
# Drop duplicate rows in the DataFrame, keeping only the first occurrence of each set of duplicated rows
df.drop_duplicates(inplace=True)
```

```python
# Select and display only the columns where the count of missing values is greater than 0
df.isnull().sum()[df.isnull().sum()>0]
```

- Before going ahead, let's remove the rows with missing values in the `app_name`, `installs`, `minimum_installs`, and `currency` columns, as they are very less in number and will not affect our analysis.

```python
# Drop rows with missing values in specified columns ('app_name', 'installs', 'minimum_installs', 'currency')
# The inplace=True parameter modifies the DataFrame in place
df.dropna(subset=['app_name', 'installs', 'minimum_installs', 'currency'], inplace=True)
```

```python
# Select and display only the columns where the count of missing values is greater than 0
df.isnull().sum()[df.isnull().sum()>0]
```

- For `Rating` we will replace the respective missing values of Rating, with respect of the average rating of their 'Category'

```python
# Calculate the average rating in each 'Category'
df.groupby('category')['rating'].mean()
```

```python
# Calculate the average rating in each 'Category'
average_ratings = df.groupby('category')['rating'].transform('mean')

# Replace missing values in 'Rating' with the respective average ratings of their Category
df['rating'] = df['rating'].fillna(average_ratings)
```

- For `rating_count` we will replace the respective missing values of rating_count, with respect of the average rating count of their 'Category'

```python
# Calculate the average rating in each 'Category'
average_rating_count = df.groupby('category')['rating_count'].transform('mean')

# Replace missing values in 'Rating' with the respective average ratings of their Category
df['rating_count'] = df['rating_count'].fillna(average_rating_count)
```

- For `size_in_KB` we will replace the respective missing values of size_in_KB, with respect of the average size of their 'Category'

```python
# Calculate the average rating in each 'Category'
average_size_KB = df.groupby('category')['size_in_KB'].transform('mean')
average_size_MB = df.groupby('category')['size_in_MB'].transform('mean')

# Replace missing values in 'Rating' with the respective average ratings of their Category
df['size_in_KB'] = df['size_in_KB'].fillna(average_size_KB)
df['size_in_MB'] = df['size_in_MB'].fillna(average_size_MB)
```

```python
# Select and display only the columns where the count of missing values is greater than 0
df.isnull().sum()[df.isnull().sum()>0]
```
